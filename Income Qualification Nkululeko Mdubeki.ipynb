```python
# Value counts of target
label_counts = train_df['Target'].value_counts().sort_index()
label_counts
```

## Check whether all members of the house have the same poverty level.

```python
unique_values = train_df.groupby('idhogar')['Target'].apply(lambda x: x.nunique() == 1)
different_households = unique_values[unique_values != True]
print('There are {} households where the family members do not all have the same poverty level.'.format(len(different_households)))
```

 ## Check if there is a house without a family head.


```python
check_head = train_df.groupby('idhogar')['parentesco1'].sum()

#Check for head
no_head = train_df.loc[train_df['idhogar'].isin(check_head[check_head == 0].index), :]
print('There are {} households without a head.'.format(no_head['idhogar'].nunique()))
```

## Set the poverty level of the members and the head of the house same in a family.

```python
for each_household in different_households.index:
    
    #find the correct label
    true_target = int(train_df[(train_df['idhogar'] == each_household) & (train_df['parentesco1'] == 1.0)]['Target'])
    
    #assign the correct label for each member
    train_df.loc[train_df['idhogar'] == each_household, 'Target'] = true_target
    
unique_values = train_df.groupby('idhogar')['Target'].apply(lambda x: x.nunique() == 1)
different_households = unique_values[unique_values != True]
print('There are {} households where the family members do not all have the same target.'.format(len(different_households)))
```
